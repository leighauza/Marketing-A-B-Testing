```python
# Note: Imports and data cleaning code for the following code cell are omitted for presentation purposes.

# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loads and Views Dataset 
df = pd.read_csv('marketing_AB.csv')

# Checks duplicates and drops unwanted columns
df.duplicated(subset='user id').sum()
df.drop(["Unnamed: 0", "user id"], axis=1, inplace=True)

# Filters categorical data
df_category = df[['test group', 'converted', 'total ads', 'most ads day', 'most ads hour']]

# Conversion Test Group
conversion_test_group = pd.crosstab(df_category['test group'], df_category['converted'], normalize='index')
print(conversion_test_group)

# Plot
conversion_test_group.plot.barh(stacked=True)

plt.title('Conversion Percentage by Test Group')
plt.xlabel('Percentage')
plt.ylabel('Test Group')
plt.legend(title='Converted', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# Conversion Day
conversion_day = pd.crosstab(df_category['most ads day'], df_category['converted'], normalize='index')
conversion_day = conversion_day.sort_values(by=True, ascending=False)
print(conversion_day)

# Plot
conversion_day = conversion_day[::-1]
conversion_day.plot.barh(stacked=True)

plt.title('Conversion Percentage Distribution by Day of the Week')
plt.xlabel('Percentage')
plt.ylabel('Day')
plt.legend(title='Converted', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# Conversion Hour
conversion_hour = pd.crosstab(df_category['most ads hour'], df_category['converted'], normalize='index')
conversion_hour = conversion_hour.sort_values(by=True, ascending=False)
print(conversion_hour)

# Plot
conversion_hour = conversion_hour[::-1]
conversion_hour.plot.barh(stacked=True, figsize=(8, 5))

plt.title('Conversion Percentage Distribution by Hour')
plt.xlabel('Percentage')
plt.ylabel('Hour')
plt.legend(title='Converted', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
