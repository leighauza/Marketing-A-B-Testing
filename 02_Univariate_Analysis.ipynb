```python
# Note: Imports and data cleaning code for the following code cell are omitted for presentation purposes.

# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loads and Views Dataset 
df = pd.read_csv('marketing_AB.csv')

# Checks duplicates and drops unwanted columns
df.duplicated(subset='user id').sum()
df.drop(["Unnamed: 0", "user id"], axis=1, inplace=True)

# Filters categorical data
df_category = df[['test group', 'converted', 'total ads', 'most ads day', 'most ads hour']]

# Plots
plt.figure(figsize=(6,4))

# Count Plot
plt.subplot(1,2,1)
sns.countplot(x=df_category['converted'], data=df_category)

plt.title(f'Count Plot - Converted')
plt.xlabel('Converted')
plt.ylabel('Count')

# Pie Plot
plt.subplot(1,2,2)
counts = df_category['converted'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%0.2f%%')

plt.title(f'Pie Plot - Converted')

# Adjust Layout
plt.tight_layout()
plt.show()
```

```python
# Plot

plt.figure(figsize=(6,4))
sns.countplot(y=df_category['most ads day'], data=df_category, order=df_category['most ads day'].value_counts().index)

plt.title(f'Count Plot - Most Ads Day')
plt.xlabel('Count')
plt.ylabel('')

plt.tight_layout
plt.show()
```

```python
# Plot
plt.figure(figsize=(6,4))
sns.countplot(y=df_category['most ads hour'], data=df_category, order=df_category['most ads hour'].value_counts().index)

plt.title(f'Count Plot - Most Ads Hour')
plt.xlabel('Count')
plt.ylabel('')

plt.tight_layout
plt.show()
```

```python
# Plot
plt.figure(figsize=(6,4))

# Histogram
plt.subplot(1,2,1)
sns.histplot(x='total ads', data=df_category[df_category['total ads'] < 50])

plt.title(f'Histogram - Total Ads')
plt.xlabel('Number of Ads')

# Boxplot
plt.subplot(1,2,2)
sns.boxplot(y='total ads', data=df_category[df_category['total ads'] < 50])

plt.title(f'Boxplot - Total Ads')
plt.ylabel('Number of Ads')

# Adjust Layout
plt.tight_layout()
plt.show()
```
